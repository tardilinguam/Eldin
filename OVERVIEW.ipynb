# Visualización Resumen de los datos

```python
import os
import matplotlib
import numpy as np
import pandas as pd
import statsmodels.api as sm  
import matplotlib.pyplot as plt


matplotlib.style.use('ggplot')
%matplotlib inline

data_path = 'data/ts'
files = sorted(os.listdir(data_path))

# Obtain product names
df = pd.read_csv(data_path + '/{0}'.format(files[-1]), skiprows=5)
names = df[df.columns[12]].tolist()
unique_names = map(lambda x: x[1] + "_" + str(names[:x[0]].count(x[1]) + 1) if names.count(x[1]) > 1 else x[1], enumerate(names))
unique_names

df = pd.DataFrame()
for f in files:
    df[f[0:-4]] = pd.read_csv(data_path + '/{0}'.format(f), skiprows=5)['Precio promedio']
for f in files:
    df[f[0:-4]] = pd.read_csv(data_path + '/{0}'.format(f), skiprows=5)['Precio promedio']

df = df.transpose()
df.columns = unique_names
df.index = pd.Index(sm.tsa.datetools.dates_from_range('2011m1', '2016m11'))








```
